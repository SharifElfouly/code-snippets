```python
from efficientnet_pytorch import EfficientNet
import torchvision.models as models
from torchsummary import summary
```

```python
def count_parameters(model):
    n_p = sum(p.numel() for p in model.parameters() if p.requires_grad) / 1000000
    print(f'{n_p} million params')
```

```python
en = EfficientNet.from_pretrained('efficientnet-b0').cuda()
```

```python
summary(en, (3, 224, 224))
```

```python
resnet18 = models.resnet18().cuda()
```

```python
count_parameters(resnet18)
```

```python
summary(resnet18, (3, 224, 224))
```
