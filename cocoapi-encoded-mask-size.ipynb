```python
from pycocotools import mask as m
import numpy as np
import sys
```

```python
SZ = (850, 450)
```

## Worst-Case

This is a worst-case approximation.

```python
w_c_m = np.asfortranarray(np.random.choice([0,1], size=SZ).astype(np.uint8))
```

```python
w_c_e_m = m.encode(w_c_m)['counts']
```

```python
w_c_e_m
```

```python
w_c_bytes = sys.getsizeof(w_c_e_m)
w_c_bytes
```

```python
print(f'1    Object: {1    * w_c_bytes / 1e6} MB')
print(f'10   Object: {10   * w_c_bytes / 1e6} MB')
print(f'100  Object: {100  * w_c_bytes / 1e6} MB')
print(f'1000 Object: {1000 * w_c_bytes / 1e6} MB')
```

## Best-Case

```python
b_c_m = np.asfortranarray(np.ones(SZ).astype(np.uint8))
```

```python
b_c_e_m = m.encode(b_c_m)['counts']
```

```python
b_c_e_m
```

```python
b_c_bytes = sys.getsizeof(b_c_e_m)
b_c_bytes
```

```python
print(f'1    Object: {1    * b_c_bytes / 1e6} MB')
print(f'10   Object: {10   * b_c_bytes / 1e6} MB')
print(f'100  Object: {100  * b_c_bytes / 1e6} MB')
print(f'1000 Object: {1000 * b_c_bytes / 1e6} MB')
```
