```python
from pydarknet import Detector, Image
from PIL import Image as Im
import numpy as np
import time
```

```python
net = Detector(bytes('/home/sharif/Documents/yolo-data/cfg/yolov3.cfg', encoding="utf-8"),
               bytes('/home/sharif/Documents/yolo-data/weights/yolov3.weights', encoding="utf-8"),
               0,
               bytes('/home/sharif/Documents/yolo-data/cfg/coco.data', encoding="utf-8"))
```

```python
net
```

```python
img = Im.open('/home/sharif/Downloads/public_place.jpg')
img = np.array(img.resize((800,450)))
img.shape
```

```python
img_darknet = Image(img)
```

```python
for _ in range(20):
    s = time.time()
    results = net.detect(img_darknet)
    e = time.time()
    print(f'FPS: {1 / (e-s)}')
```
